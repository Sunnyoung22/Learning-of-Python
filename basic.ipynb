```python
#创建一list，变量名为a
a = [[1, 2, 3], [4, 5, 6]]
print(a)
```

```python
#引入numpy库,调用array函数生成一个矩阵
import numpy as np
a = np.array(a)
print(a)
```

```python
#matplotlib inline是IPython的魔法函数，可以在IPython编译器里直接使用，作用是内嵌画图
#导入Python绘图包
%matplotlib inline 
from matplotlib import pyplot as plt
```

```python
#显示图像
plt.plot(a)
plt.show()
```
